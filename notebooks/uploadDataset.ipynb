```python
import os
import json
import random
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm
from datasetCleaner import get_base_id
```

```python
DATA_PATH = "../data/train"
IMAGES_PATH = os.path.join(DATA_PATH, "images")
LABELS_PATH = os.path.join(DATA_PATH, "labels")
TARGETS_PATH = os.path.join(DATA_PATH, "targets")

SEED = 42
random.seed(SEED)
```

```python
files = os.listdir(IMAGES_PATH)

print("Total archivos:", len(files))
print(files[:10])
```

```python
ids = set()

for f in files:
    if "_pre_disaster" in f:
        ids.add(get_base_id(f))

ids = list(ids)
print("Total escenas:", len(ids))
```

```python
random.shuffle(ids)

n = len(ids)

train_ids = ids[:int(0.7*n)]
val_ids   = ids[int(0.7*n):int(0.85*n)]
test_ids  = ids[int(0.85*n):]

print(len(train_ids), len(val_ids), len(test_ids))
```

```python
def save_list(path, data):
    with open(path, "w") as f:
        for item in data:
            f.write(item + "\n")

save_list("../data/train.txt", train_ids)
save_list("../data/val.txt", val_ids)
save_list("../data/test.txt", test_ids)
```
